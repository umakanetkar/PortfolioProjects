Importing stop words from Python's NLTK library

```python
import nltk
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
```

Importing the Regex package

```python
import re
```

Opening the text file 

```python
text= open('Trump_Raw_Tweets_Fall22.txt', 'r')
```

To clean the file, I am encoding and then decoding them. Then I am using regex, to get rid of all http links, "@" meantions and the ampersand characters (&amp) and hashtag symbols

```python
for files in text:
    files = files.encode(errors='ignore').decode('utf8')
    files = re.sub(r'\\x', ' ', files)
    files = re.sub(r"http\S+", " ", files)
    files = re.sub("@[A-Za-z0-9_]+"," ", files)
    files = re.sub("&[A-Za-z0-9_]+"," ", files)
    files = re.sub("[^A-Za-z.]"," ", files)
    
    files = re.sub("RT", "", files)
    files = re.sub(r'\b\w\b', ' ', files)
    print(files)
```

Opening the textfiles containing positive and negative words

```python
pos_text = open('positive.txt', 'r')
```

```python
neg_text= open('negative.txt', 'r')
```

Creating new variables to track count of positive/negative/stop words and initially assigning them 0

```python
pos_count=0
neg_count=0
stop_count=0
others_count=0
```

Forming empty lists for positive, negative and stop words and appending them with all words from their respective text files so that I can check presence of those words in the twitter file. Using strip function to remove trailing characters. 

```python
pos_list=[]
for pos in pos_text:
    pos=pos.strip()
    pos_list.append(pos)
```

```python
neg_list=[]
for neg in neg_text:
    neg=neg.strip()
    neg_list.append(neg)
```

```python
stop_list=[]
for stop_word in stops:
    stop_word=stop_word.strip()
    stop_list.append(stop_word)
```

Using split function and making count of all words.

```python
split_text=files.split()
for words in split_text:
    if words.lower() in pos_list:
        pos_count+=1
    if words.lower() in neg_list:
        neg_count-=1
    if words.lower() in stop_list:
        stop_count+=1
    else:
        others_count+=1
        
total_count= pos_count-neg_count+stop_count+others_count        
```

```python
print(" The positive word count is %s" %pos_count)
print(" The negative word count is %s" %neg_count)
print(" The stop word count is %s" %stop_count)
print(" The other word count is %s" %others_count)
print(" The total word count is %s" %total_count)
```

What’s the ratio of positive/negative/stop word/others compare to the total word count?

```python
print("The ratio of positive words to total word count is %s ." %(pos_count/total_count))
print("The ratio of negative words to total word count is %s ." %(-neg_count/total_count))
print("The ratio of stop words to total word count is %s ." %(stop_count/total_count))
print("The ratio of other words to total word count is %s ." %(others_count/total_count))
```

What’s the ratio for positive vs negative word count? 

```python
print("The ratio of positive words to negative words is %s" %(-pos_count/neg_count))
```
